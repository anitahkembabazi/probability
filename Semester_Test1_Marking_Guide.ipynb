**Q6: The Probability Mass Function (PMF) of the binomial distribution is calculated using the a particular mathematical equation. Explain the purpose of the each parameter within the equation [4 MARKS]**

- The equation is: 
- P(X = x) = ( n x )px (1 - p) n -  x

- p - probability of success on a single trial. The chances or likelihood that the event will occur in one trial. For example, with a fair or two sided coin, the chance of flipping a head on a single flip is 50% or 0.5. 
- n represents to the total number of trials. It represents the total number of times an experiment or event is carried out. For example, if we flip a coin five times, each flip is trial giving us n = 5. 
- x = the number of successful trials. It represents the number of times the desired outcome occurs.  For example, if we flip a coin 5 times and we want to know the probability of getting 2 heads in those 5 flips, k =2. 
- (n x) = the binomial coefficient. It calculates the number of ways to choose "x" successes out of "n" trials. If we want 2 heads out of 5 flips, this will give is  all the possible combinations to get 2 heads from the 5 coin flips. 
- P(X = x) represents a specific probability in the binomial distribution. It basically states the question..."What is the probability of getting x  successes in a given number of trials?"
- (1 - p) is the chance that we don't succeed on a single try.

**Q10a:Discuss the null hypotheses of the following statistical tests [6 MARKS]**
- i. Shapiro-Wilks test:
The null hypothesis states that theres no difference between the collected datas distribution and the expected normal distribution of a data model. If our threshold or alpha=0.05 or 5%, then at p>0.05 we cannot reject the null hypothesis (i.e. data is normally distributed), at p<0.05, we reject the null hypothesis (i.e. data is not normally distributed).

- ii. One-way ANOVA test:
The null hypothesis states that theres no difference in the sample means of variables being compared (i.e. at p>0.05 we cannot reject the null hypothesis and that two variables have no statistically significant relationship or no effect on each other).. 

- iii. Pearsons correlation test:
The null hypothesis states that the correlation coefficient is not different from zero (i.e. r=0, implying no linear relationship between the variables)

**Q11c: What is the relationship between the variables "Target" and "Mother's qualification?" [20 MARKS]**

```python
Q11['Target'].unique()
```

There are 3 unique target categories (Drouput, Graduate, Enrolled).

```python
#Generate a contigency table for the two variables
crosstab = pd.crosstab(index = Q11['Mother\'s qualification'], columns= Q11['Target'])

#Use the contigeny table to generate a barplot
%matplotlib inline
crosstab.plot.bar()

#Add title and axes labels to barplot
plt.xlabel('Mother\'squalification')
plt.ylabel('Count')
plt.title('Relationship between Qualitfications of a Mother and the Study Outcome of a Student')
```

```python
#Check the statistical relationship
chi2_stat, p, dof, expected = scipy.stats.chi2_contingency(crosstab)
print(f"chi2 statistic:     {chi2_stat:.5g}")
print(f"p-value:            {p:.5g}")
print(f"degrees of freedom: {dof}")
print("expected frequencies:\n",expected)
```

            Result:
* The graph shows that the counts of students who are enrolled, dropouts and graduates varies for the qualifications of mothers.
* For instance, Qualification 37 has the highest number of student dropouts, Qualification 1 has the highest number of enrolled students and graduate students.
* The Chi-Square statistic ranges from zero to infinity. The more associated two variables are, the larger the Chi-Square statistic will be
* Therefore, there's a strong statistical correlation between the variables Target and Mother's qualification (chi2 = 217.95, p =  5.8238e-21) implying the counts of the varying outcomes of the students are correlated to the qualification of the  mothers.
