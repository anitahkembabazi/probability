# 1. Simple random sampling

**A. Implement random sampling on a population using visuals**

```python
#Import libraries
import numpy as np
import pandas as pd
 
# Creation of a sample population
population = pd.DataFrame({
    'ID': range(1, 101),  # Unique identifiers for individuals
    'Age': np.random.randint(18, 70, size=100),  # Random age between 18 and 70 years
    'Gender': np.random.choice(['Male', 'Female'], size=100),  # Random gender
    'Income': np.random.normal(50000, 10000, size=100)  # Random income with mean 50000 and standard deviation 10000
})
#View population
population
```

```python
#Import visualisation libraries
import matplotlib.pyplot as plt
import seaborn as sns
 
# Histogram plot to visualize the age distribution
plt.figure(figsize=(10, 6))
sns.histplot(population['Age'], bins=10, kde=True, color='skyblue')
plt.title('Age Distribution in the Population')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#Randomly sample 20 individuals from the population
#Define a function to do that
def simple_random_sampling(population, n):
    return population.sample(n)
 
# Use the function to randomly sample 20 individuals
random_sample = simple_random_sampling(population, 20)
print("Simple Random Sample:")
print(random_sample)
```

```python
#Generate visuals that compare the age distribution of the random sample and the population
import matplotlib.pyplot as plt
import seaborn as sns
 
# Overlay histogram plot to visualize the age distribution in the population and the random sample
plt.figure(figsize=(10, 6))
 
# Age distribution in the population
sns.histplot(population['Age'], bins=20, kde=True, color='skyblue', label='Population')
 
# Age distribution in the random sample
sns.histplot(random_sample['Age'], bins=10, kde=True, color='salmon', label='Random Sample')
 
plt.title('Age Distribution: Population vs Random Sample')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
 
plt.show()
```

- We notice that the random sample doesn't show representation for some ages e.g. 57-59, 65-78
- So our sampling is biased
- What happens if we increase the sample size?

```python
# Use the function to randomly sample 40 individuals
random_sample2 = simple_random_sampling(population, 40)
print("Simple Random Sample of 40:")
print(random_sample2)

## Overlay histogram plot to visualize the age distribution in the population and the random sample of 40
plt.figure(figsize=(10, 6))
 
# Age distribution in the population
sns.histplot(population['Age'], bins=20, kde=True, color='skyblue', label='Population')
 
# Age distribution in the random sample
sns.histplot(random_sample2['Age'], bins=10, kde=True, color='salmon', label='Random Sample of 40')
 
plt.title('Age Distribution: Population vs Random Sample of 40')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
 
plt.show()
```

- Notice that the bias lessens with an increase in sampling

# 2. Stratified Sampling
- Divides the population into homogeneous groups called â€œstrataâ€, and then selects a random sample from each of these strata. 
- This method is used when the population has heterogeneity/diversity in some key characteristics, and the objective is to ensure that the sample accurately reflects this heterogeneity.

```python
#For instance using age as the stratum and randomly choosing different percentages of customers of different ages
# Define a function for stratified sampling by Age
def stratified_sampling(data, n, stratification):
    sample = pd.DataFrame()
    for value, proportion in stratification.items():
        stratum = data[data['Customer_Age'] == value]
        if stratum.empty:  # Check if there are individuals in the given stratum
            print("There are no individuals in the population with age ", value)
            print("Modify the population or change the age.")
            return sample
        else:
            # Check if there are enough individuals in the given stratum
            if len(stratum) < int(n * proportion):
                print("There are not enough individuals in the population with age ", value)
                return sample
            else:
                stratum_sample = stratum.sample(int(n * proportion))
                sample = pd.concat([sample, stratum_sample])
    return sample
```

```python
# Example of using stratified sampling by age to select varying percentages of customers at different ages
#We want to sample 1000 customers but take different percentages for each age
#E.g. customers of age 20 (sampling 20% of population), age 35(sampling 30% of population) etc
age_stratification = {20: 0.2, 35: 0.3, 40: 0.2, 50: 0.1, 60: 0.2}  # Proportions for each age stratum 
stratified_age_sample = stratified_sampling(data, 1000, age_stratification) # sampling in total 1000 
print("\nStratified Sampling by Age:")
print(stratified_age_sample)
```

- It is important to have prior knowledge about the dataset when doing stratified sampling (i.e. knowing which age customers are)

```python
import matplotlib.pyplot as plt
 
# Calculate proportions in the population
population_proportions = {age: len(data[data['Customer_Age'] == age]) / len(data) for age in age_stratification.keys()}
 
# Calculate proportions in the sample
sample_proportions = {age: len(stratified_age_sample[stratified_age_sample['Customer_Age'] == age]) / len(stratified_age_sample) for age in age_stratification.keys()}
 
# Ages and proportion values
ages = list(age_stratification.keys())
population_props = list(population_proportions.values())
sample_props = list(sample_proportions.values())
 
# Bar width
bar_width = 0.35
 
# Bar positions on the plot
positions = range(len(ages))
 
# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(positions, population_props, bar_width, label='Population', color='skyblue')
plt.bar([p + bar_width for p in positions], sample_props, bar_width, label='Sample', color='salmon')
 
# Labels and title
plt.xlabel('Age')
plt.ylabel('Proportion')
plt.title('Comparison of Proportions between Population and Sample by Age')
plt.xticks([p + bar_width / 2 for p in positions], ages)
plt.legend()
 
plt.show()
```

# 3. Cluster Sampling

- A form of probability sampling for larger populations that are spatially stratified/divided
- Population is divided into known smaller groups/clusters
- Clusters can be geographically-based e.g. districts, parishes
- Using the Bike_Sales dataset, we could stratify the customers by the country they come from


```python
#Let's see how many countries are in the dataframe
data['Country'].nunique() #There are 6 countries
```

```python
#Import libraries
import pandas as pd
import random
#set a random seed
random.seed(1)

#select 2 random countries as the clusters (k=4)
selected_clusters = random.sample(data['Country'].tolist(), k=2)
print(selected_clusters)
```

```python
#select all customers from the selected countries
sampled_data = data[data['Country'].isin(selected_clusters)]
print(sampled_data)
```

```python
#Now let's compare the age distribution of the sampled clusters against the population age
## Overlay histogram plot to visualize 
plt.figure(figsize=(10, 6))
 
# Age distribution in the population
sns.histplot(data['Customer_Age'], bins=20, kde=True, color='magenta', label='Population')
 
# Age distribution in the random sample
sns.histplot(sampled_data['Customer_Age'], bins=10, kde=True, color='green', label='Clustered sample')
 
plt.title('Age Distribution: Population vs Clustered Sample of 4 countries')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
 
plt.show()
```
