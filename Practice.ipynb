```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
disease=pd.read_csv("Assignment 4_Remedial.csv")
disease.head()
```

```python
disease.dtypes
```

```python
#Visual relationship between Continuous variables(Sleeptime and BMI)
disease.plot.scatter(x='SleepTime',y='BMI',figsize=(10,5),title='SleepTime'+'VS'+'BMI')
```

```python
#relationship test between two continuous variables
Groupeddata=disease[['BMI','SleepTime']]
Groupeddata.corr()
```

```python
#Visual relationship between two categorical variables(Diabetic and PhysicalActivity)
disease_crosstable=pd.crosstab(index=disease['Diabetic'],columns=disease['PhysicalActivity'])
disease_crosstable
```

```python
#chi-test for two categorical
from scipy.stats import chi2_contingency
Chitestresults=chi2_contingency(disease_crosstable)
Chitestresults[1]
```

```python
#bar plot
disease_crosstable.plot.bar()
```

```python
#visual relation ship for categorical and continuous variables(BMI and PhysicalActivity)
disease.boxplot(column='BMI',by='PhysicalActivity',figsize=(18,10))
```
