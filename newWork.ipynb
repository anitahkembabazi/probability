#Generating statistical tests 

statistical tests for two continuous variables using the Pearson Correlation coefficient

```python
%pip install scipy 
```

```python
import scipy.stats as stats
```

```python
#pearson correlation test between unit cost and profit
corr, p_value = stats.pearsonr(data['Unit_Cost'], data['Profit'])

print(f'Pearson correlation coefficient: {corr}')
print(f'p-value: {p_value}')
```

Pearson correlation coefficient is  0.7883307282059594 which indicates a strong positive correlation between unit cost and profit
p-value: 0.0

statistical test between two categorical variables

```python
%pip install numpy
```

```python
import numpy as np
```

```python
#creating a contingency table
contingency_table = pd.crosstab(data['Age_Group'], data['Year'])
print("Contingency Table:")
print(contingency_table)
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
#Performing the chi square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"\nChi-square Statistic: {chi2:.2f}")
print(f"P-value: {p:.4f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)
```

```python
alpha = 0.05  # significance level
if p < alpha:
    print("Reject the null hypothesis: There is a significant association between Age_Group and year.")
else:
    print("Fail to reject the null hypothesis: There is no significant association between Age_Group and year.")
```

```python
#tests between one categorical and one continuous variable
import pandas as pd
from scipy.stats import f_oneway
```

```python
groups = [data[data['Age_Group'] == group]['Profit'] for group in data['Age_Group'].unique()]
```

```python
f_stat, p_value = f_oneway(*groups)
print(f"ANOVA F-Statistic: {f_stat:.2f}")
print(f"P-value: {p_value:.4f}")
```

ANOVA F-Statistic: 57.57
P-value: 0.0000

```python
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: At least one group mean is significantly different.")
else:
    print("Fail to reject the null hypothesis: All group means are not significantly different.")
```

Reject the null hypothesis: At least one group mean is significantly different.
