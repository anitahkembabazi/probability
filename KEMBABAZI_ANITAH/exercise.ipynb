```python
import pandas as pd
```

```python
import numpy as np
```

```python
data = pd.read_excel("diamonds_new.xlsx")
data
```

```python
data.info()
```

```python
from scipy import stats
```

```python
# number one.
# Step 1: Check the unique t
table_cat= data['table'].unique()

grouped_prices = [data[data['table'] == table]['price'] for table in table_cat]

# Step 3: Perform the ANOVA test
anova_result = stats.f_oneway(*grouped_prices)

# Step 4: Print the results
print(f"ANOVA F-statistic: {anova_result.statistic}")
print(f"ANOVA p-value: {anova_result.pvalue}")

alpha = 0.05
if anova_result.pvalue < alpha:
    print("Reject the null hypothesis. There is a significant difference in the average prices across table categories.")
else:
    print("Fail to reject the null hypothesis. The average prices of all table categories are the same.")
```

```python
#number two
premium_diamonds = data[data['cut'] == 'Premium']

sample_mean_premium = np.mean(premium_diamonds['price'])
sample_size_premium = len(premium_diamonds)

# Step 3
population_std = np.std(premium_diamonds['price'], ddof=1)  # Using sample standard deviation

# Step 4: Perform the Z-test
population_mean = 150  # 
z_score = (sample_mean_premium - population_mean) / (population_std / np.sqrt(sample_size_premium))

p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Print results
print(f"Sample Mean (Premium Diamonds): {sample_mean_premium}")
print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")

# Decision based on p-value
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The average price of Premium diamonds is significantly different from 150.")
else:
    print("Fail to reject the null hypothesis. The average price of Premium diamonds is not significantly different from 150.")
```
